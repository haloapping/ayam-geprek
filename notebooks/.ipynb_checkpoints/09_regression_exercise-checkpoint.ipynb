# Import Modules

```python
import pandas as pd
import seaborn as sns
import torch

from time import time
from torch import nn, optim
from torchmetrics import R2Score

device = "cuda" if torch.cuda.is_available else "cpu"
```

# Import dataset

```python
california_houses_df = pd.read_csv("../datasets/california_houses.csv")
california_houses_df.head()
```

# Split feature and target

```python
y = california_houses_df.pop("Median_House_Value").values.reshape(-1, 1)
X = california_houses_df.values

print(f"X shape: {X.shape}")
print(f"y shape: {y.shape}")
```

# Split train and test dataset

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.3)
print(f"X_train shape: {X_train.shape}")
print(f"y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}")
print(f"y_test shape: {y_test.shape}")
```

```python
X_train = torch.FloatTensor(X_train)
y_train = torch.FloatTensor(y_train)

X_test = torch.FloatTensor(X_test)
y_test = torch.FloatTensor(y_test)
```

# Build Architecture

## TensorDataset and DataLoader

```python
from torch.utils.data import TensorDataset, DataLoader

batch_size = 128

# TensorDataset
tensor_dataset_train = TensorDataset(X_train, y_train)
tensor_dataset_test = TensorDataset(X_test, y_test)

# DataLoader
train_dataloader = DataLoader(tensor_dataset_train, batch_size=batch_size)
test_dataloader = DataLoader(tensor_dataset_test, batch_size=batch_size)
```

## Model

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()

        self.fc = nn.Sequential(
            nn.Linear(13, 64),
            nn.Linear(64, 64),
            nn.Linear(64, 1)
        )
    
    def forward(self, x):
        x = nn.functional.normalize(x)
        x = self.fc(x)
        
        return x
    
model = Model()
```

```python
model.state_dict()
```

## Loss Function

```python
mse_loss = nn.MSELoss()
mse_loss
```

## Optimizer

```python
adamw_optimizer = optim.AdamW(model.parameters(), lr=0.1,)
adamw_optimizer
```

## Evalutaion Metric

```python
r2_score = R2Score()

r2_score
```

## Loop function

```python
def train_loop(dataloader, model, loss_fn, optimizer, eval_metric):
    print("Train:")
    
    losses = []
    for batch, (feature, label) in enumerate(dataloader):
        # Forwardpropagation
        pred_label = model(feature)
        loss = loss_fn(pred_label, label)
        acc = eval_metric(pred_label, label)
        
        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()
        
        loss = loss.item()
        losses.append(loss)
        
        print(f"Batch-{batch + 1} | R2 Score: {acc:>7f} | Loss: {loss:>7f}")
        
    acc = eval_metric.compute()
    print(f"R2 Score : {acc}")
    print(f"Loss     : {sum(losses) / len(dataloader)}")
    eval_metric.reset()
    
def test_loop(dataloader, model, loss_fn, eval_metric):
    print("\nTest:")
    
    losses = []
    for batch, (feature, label) in enumerate(dataloader):
        # Forwardpropagation
        pred_label = model(feature)
        loss = loss_fn(pred_label, label)
        acc = eval_metric(pred_label, label)
        
        with torch.no_grad():
            loss = loss.item()
            losses.append(loss)
        
        print(f"Batch-{batch + 1} | R2 Score: {acc:>7f} | Loss: {loss:>7f}")
        
    acc = eval_metric.compute()
    print(f"R2 Score : {acc}")
    print(f"Loss     : {sum(losses) / len(dataloader)}")
    eval_metric.reset()
```

```python
epochs = 100

start = time()
for epoch in range(epochs):
    print(f"EPOCH {epoch + 1}")
    print("=" * 46, end="\n")
    train_loop(train_dataloader, model, mse_loss, adamw_optimizer, r2_score)
    test_loop(test_dataloader, model, mse_loss, r2_score)
    print("=" * 46, end="\n\n")
torch.save(model.state_dict(), "../pretrained_model/house_price.pth")
stop = time()

total_time = stop - start
print(f"Training time: {(total_time / 60):.3f} minutes")
```
